```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
from scipy.io import loadmat
data1 = loadmat('ex6data1.mat')
```

```python
X1, y1 = data1['X'], data1['y']
```

```python
m1, n1 = X1.shape[0], X1.shape[1]
```

```python
pos1, neg1 = (y1==1).reshape(m1, 1), (y1==0).reshape(m1, 1)
```

```python
plt.scatter(X1[pos1[:, 0], 0], X1[pos1[:, 0], 1], color='red', marker='+', s=50)
plt.scatter(X1[neg1[:, 0], 0], X1[neg1[:, 0], 1], color='blue', marker='x', s=50)
plt.show()
```

```python
from sklearn.svm import SVC
model1 = SVC(kernel='linear')
model1.fit(X1, y1.flatten())
```

```python
score1 = accuracy_score(y1, model1.predict(X1))
score1
```

```python
data2 = loadmat('ex6data2.mat')
X2, y2 = data2['X'], data2['y']
```

```python
m2, n2 = X2.shape[0], X2.shape[1]
```

```python
plt.figure(figsize=(7.5, 7.5))
pos2, neg2 = (y2==1).reshape(m2, 1), (y2==0).reshape(m2, 1)
plt.scatter(X2[pos2[:, 0], 0], X2[pos2[:, 0], 1], c='red', marker='x', s=50)
plt.scatter(X2[neg2[:, 0], 0], X2[neg2[:, 0], 1], c='blue', marker='+', s=50)
plt.show()
```

```python
model2 = SVC(kernel='rbf', gamma=30)
model2.fit(X2, y2.flatten())
```

```python
score2 = accuracy_score(y2, model2.predict(X2))
score2
```
